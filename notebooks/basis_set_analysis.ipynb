```python
def get_sigma_r_vals(exponents):
    nfuncs = {'S':1, 'P':3, 'D':6, 'F':10, 'G':15, 'H':21}
    lvals = {'S':0, 'P':1, 'D':2, 'F':3, 'G':4, 'H':5}
    r_vals = {}
    alpha_inv = {}
    for function in exponents:
        r_vals[function] = [ np.sqrt(np.log(2)/exp) for exp in exponents[function]]
    return r_vals, alpha_inv
```

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
data_rvals = {}
data_alpha_inv = {}
for basis in exponents:
    rvals, alpha_inv = get_sigma_r_vals(exponents[basis])
    rvals_list = [values[i] for values in rvals.values() for i in range(len(values))]
    alpha_inv_list = [values[i] for values in alpha_inv.values() for i in range(len(values))]
    print(basis, len(rvals_list))
    data_rvals[basis] = rvals_list
    data_alpha_inv[basis] = alpha_inv_list
```

# Keldysh Parameters

```python
import sys
import os
sys.path.append(os.path.join(os.path.realpath('../'),'scripts'))
from keldysh import *
```

# Helium

```python
plot_keldysh('He', I_lim=[2.5E13,5.1e14], wln_lim=[350, 1800], plotname='he')
```

```python
print('He')
print('I0       gamma    Up   Ecutoff  ncut   Rmax')
w0 = wavelength_to_freq(800e-9)
for I in [2E14,3E14,5E14]:
    data_dict = keldysh(I, 800, 'He')
    gamma = data_dict['gamma']
    Up = data_dict['Up']
    Ecutoff = data_dict['Ecutoff']
    Rmax = data_dict['Rmax'] 
    ncut = int(Ecutoff/w0)
    print('{I:1.1E}, {gamma:3.2f}, {Up:1.4f}, {Ecutoff:3.4f},  {ncut:3d},   {Rmax:3.4f}'.format(
        I=I,gamma=gamma, Up=Up, Ecutoff=Ecutoff, ncut=ncut, Rmax=Rmax))
```
